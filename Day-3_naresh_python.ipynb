```python
x=x.reshape(30,1)
```

```python
x  # input data should be in 2d 
```

#### import the linear regression class

```python
from sklearn.linear_model import LinearRegression
```

```python
### create the objec of linear regression
```

```python
regressor = LinearRegression()
```

```python
## Train the algorithm
```

```python
regressor.fit(x,y)
```

```python
### predict on the unseen data
```
